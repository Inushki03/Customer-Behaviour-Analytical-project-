```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/customer_shopping_behavior.csv")
df.head()
```

```python
df.isna().sum()
```

Take each group (x), find its median, and fill its NaN values with that median

```python
df["Review Rating"]=df.groupby("Category")['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create age groups
labesl=["Young_Adult","Adult","Middle Age","Senior"]
df["age_group"]=pd.qcut(df['age'],q=4,labels=labesl)
```

```python
df[['age','age_group']].head(5)
```

```python
df["frequency_of_purchases"].unique()
```

```python
#create new column frequency_purchases
freequncy_purchases={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
df['frequency_purchases']= df['frequency_of_purchases'].map(freequncy_purchases)
```

```python
df[['frequency_purchases','frequency_of_purchases']].head(10)
```
